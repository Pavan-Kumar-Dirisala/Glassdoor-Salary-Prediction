## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Drop the index column if not needed
data.drop(columns=['Unnamed: 0'], inplace=True)


````

````python
import pandas as pd

# Step 1: Replace '-1' with NaN for easier handling (optional if you're tracking separately)
# data['Salary Estimate'] = data['Salary Estimate'].replace('-1', np.nan)

# Step 2: Filter only rows that contain salary ranges like "$XXK-$YYK"
pattern = r'^\$\d+[Kk]\s*-\s*\$\d+[Kk]'
valid_salary = data['Salary Estimate'].str.contains(pattern, regex=True, na=False)
salary_data = data[valid_salary].copy()

# Step 3: Remove extra text like "Glassdoor est.", "Employer Provided Salary:"
salary_data['Salary Estimate Cleaned'] = salary_data['Salary Estimate'].str.replace(r'\(.*?\)', '', regex=True)
salary_data['Salary Estimate Cleaned'] = salary_data['Salary Estimate Cleaned'].str.replace(r'[^\d\-]', '', regex=True)

# Step 4: Split into min and max salary
salary_data[['min_salary', 'max_salary']] = salary_data['Salary Estimate Cleaned'].str.split('-', expand=True)

# Step 5: Strip and convert to integers
salary_data['min_salary'] = salary_data['min_salary'].astype(int)
salary_data['max_salary'] = salary_data['max_salary'].astype(int)

# Step 6: Calculate average salary
salary_data['avg_salary'] = (salary_data['min_salary'] + salary_data['max_salary']) // 2

# Step 7: Merge the new salary columns back to original data (if needed)
data = data.merge(
    salary_data[['min_salary', 'max_salary', 'avg_salary']],
    left_index=True,
    right_index=True,
    how='left'
)
````

#### What all missing value imputation techniques have you used and why did you use those techniques?

To handle missing values in the dataset, I applied different imputation techniques based on the type and role of each column. For categorical columns like `Headquarters`, `Size`, `Type of ownership`, `Industry`, `Sector`, and `Revenue`, I filled missing values with the label `'Unknown'`. This approach allowed me to retain all records without making any assumptions about the missing categories and ensured that these unknowns would be handled appropriately during one-hot encoding. For numerical columns such as `Rating` and `Founded`, I chose to fill missing values using the median, as it is less sensitive to outliers and better represents the central tendency, especially in skewed distributions. However, for the `avg_salary` column, which is the target variable, I did not apply any imputation. Instead, I dropped rows with missing salary data to avoid introducing bias into the model. This combination of strategies helped me maintain both data integrity and model reliability.


### 2. Handling Outliers

````python
# Define the function to remove outliers using IQR
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] >= lower) & (df[column] <= upper)]

# Apply IQR outlier removal to key numeric columns
for col in ['avg_salary', 'Rating', 'Company_Age']:
    data_cleaned = remove_outliers_iqr(data_cleaned, col)

# Show result
print("Data shape after removing outliers:", data_cleaned.shape)
````

##### What all outlier treatment techniques have you used and why did you use those techniques?



To handle outliers in my dataset, I primarily used the **Interquartile Range (IQR) method**. I applied this technique to the numerical columns like `avg_salary`, `Rating`, and `Company_Age`. The reason I chose IQR is because it is one of the most robust and commonly used statistical methods for detecting and removing outliers. It identifies values that lie below Q1 - 1.5*IQR or above Q3 + 1.5*IQR as outliers.

I preferred this over methods like Z-score because IQR doesn‚Äôt assume a normal distribution, which was ideal for my dataset since salary and company features often have skewed distributions. After identifying the outliers, I filtered them out to prevent their influence on downstream machine learning models.

Importantly, I made sure that outlier removal was only applied to the version of the dataset used for modeling. The visualizations were done on the original data so that they reflect the complete picture of the job market insights.

### 4. Textual Data Preprocessing

#### 1. Expand Contraction

````python
# Expand Contraction
import re
import contractions

# Apply contraction expansion to 'Job Description'
data_encoded['Job Description'] = data_encoded['Job Description'].apply(lambda x: contractions.fix(x))
````

#### 2. Lower Casing

````python
# Lower Casing
# Convert all job descriptions to lowercase
data_encoded['Job Description'] = data_encoded['Job Description'].str.lower()
````

#### 3. Removing Punctuations

````python
# Remove punctuation from job descriptions
import string
data_encoded['Job Description'] = data_encoded['Job Description'].str.replace(f"[{string.punctuation}]", " ", regex=True)
````

#### 4. Removing URLs & Removing words and digits contain digits.

````python
# Remove URLs & Remove words and digits contain digits
import re

def clean_urls_digits(text):
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Remove words containing digits
    text = re.sub(r'\w*\d\w*', '', text)
    return text

# Apply to Job Description column
data_encoded['Job Description'] = data_encoded['Job Description'].apply(clean_urls_digits)
````

#### 5. Removing Stopwords & Removing White spaces

````python
# Remove White spaces
# Remove extra whitespaces from job description
data_encoded['Job Description'] = data_encoded['Job Description'].apply(lambda x: ' '.join(x.split()))
````

#### 7. Tokenization

````python
import spacy

nlp = spacy.load("en_core_web_sm")

def spacy_tokenize(text):
    doc = nlp(text)
    return [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]

data_encoded['tokens'] = data_encoded['Job Description'].apply(spacy_tokenize)
````

````python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Step 1: Join tokens into a single string
data_encoded['clean_text'] = data_encoded['tokens'].apply(lambda x: ' '.join(x))

# Step 2: Apply TF-IDF Vectorizer on the clean_text
tfidf = TfidfVectorizer(max_features=300)  # Adjust number as needed
tfidf_matrix = tfidf.fit_transform(data_encoded['clean_text'])

# Step 3: Convert TF-IDF output to DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), 
                        columns=tfidf.get_feature_names_out(), 
                        index=data_encoded.index)

# Step 4: Drop raw text and tokens from data_encoded
dataencoded = data_encoded.drop(columns=['Job Description', 'tokens', 'clean_text'])

# Step 5: Concatenate original features with TF-IDF features
dataencoded = pd.concat([dataencoded, tfidf_df], axis=1)
````

Lemmatization, in particular, helped reduce words to their base forms, which is essential for consistent and meaningful text analysis.

````python
import pandas as pd
import numpy as np
import re
import string
import spacy
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack
import joblib

# === Step 1: Load Saved Assets ===
best_model = joblib.load("best_xgboost_model.pkl")
selector = joblib.load("select_kbest.pkl")
tfidf_vectorizer = joblib.load("tfidf_vectorizer.pkl")
onehot_columns = joblib.load("onehot_columns.pkl")  # Should be 963 columns

# === Step 2: New Job Data ===
new_data = pd.DataFrame([{
    'Job Title': 'Data Scientist',
    'Company Name': 'XYZ Corp',
    'Location': 'San Francisco, CA',
    'Headquarters': 'New York, NY',
    'Size': '51 to 200 employees',
    'Type of ownership': 'Private',
    'Industry': 'Information Technology',
    'Sector': 'Technology',
    'Revenue': '$10 to $50 million (USD)',
    'Rating': 3.8,
    'Company_Age': 10,
    'Has_Competitor': 1,
    'Job Description': '''
        Looking for a data scientist with experience in Python, SQL, and machine learning.
        Familiarity with NLP and cloud services is a plus. 
        Apply now at www.xyzjobs.com
    '''
}])

# === Step 3: Clean Job Description ===
new_data['Job Description'] = new_data['Job Description'].str.lower()
new_data['Job Description'] = new_data['Job Description'].str.replace(f"[{string.punctuation}]", " ", regex=True)

def clean_urls_digits(text):
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'\w*\d\w*', '', text)
    return text

new_data['Job Description'] = new_data['Job Description'].apply(clean_urls_digits)
new_data['Job Description'] = new_data['Job Description'].apply(lambda x: ' '.join(x.split()))

# === Step 4: Tokenization + Lemmatization ===
nlp = spacy.load("en_core_web_sm")

def spacy_tokenize(text):
    doc = nlp(text)
    return [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]

new_data['tokens'] = new_data['Job Description'].apply(spacy_tokenize)
new_data['clean_text'] = new_data['tokens'].apply(lambda x: ' '.join(x))

# === Step 5: TF-IDF Transformation (Limit to 297) ===
tfidf_features_full = tfidf_vectorizer.transform(new_data['clean_text'])

# Reduce TF-IDF features to 297 (top 297 variance)
from sklearn.feature_selection import VarianceThreshold
tfidf_dense = tfidf_features_full.toarray()
variances = tfidf_dense.var(axis=0)
top_297_idx = np.argsort(variances)[-297:]
tfidf_features = tfidf_dense[:, top_297_idx]

# === Step 6: One-Hot Encoding ===
categorical_cols = ['Job Title', 'Company Name', 'Location', 'Headquarters',
                    'Size', 'Type of ownership', 'Industry', 'Sector', 'Revenue']
base_cols = ['Rating', 'Company_Age', 'Has_Competitor']

new_data_encoded = pd.get_dummies(new_data[categorical_cols + base_cols], columns=categorical_cols, drop_first=True)

# Align columns with training
missing_cols = list(set(onehot_columns) - set(new_data_encoded.columns))
extra_cols = list(set(new_data_encoded.columns) - set(onehot_columns))

if extra_cols:
    print(f"‚ö†Ô∏è Dropping extra columns: {extra_cols}")
    new_data_encoded.drop(columns=extra_cols, inplace=True)

if missing_cols:
    missing_df = pd.DataFrame(0, index=new_data_encoded.index, columns=missing_cols)
    new_data_encoded = pd.concat([new_data_encoded, missing_df], axis=1)

new_data_encoded = new_data_encoded[onehot_columns]  # reorder

# === Step 7: Combine Features ===
from scipy.sparse import csr_matrix

tfidf_sparse = csr_matrix(tfidf_features)  # Convert dense to sparse
combined_features = hstack([tfidf_sparse, csr_matrix(new_data_encoded.values)])

# print(f"‚úÖ Combined feature shape: {combined_features.shape}")

# === Step 8: Sanity Check ===
assert combined_features.shape[1] == selector.n_features_in_, \
    f"Feature mismatch: expected {selector.n_features_in_}, got {combined_features.shape[1]}"

# === Step 9: SelectKBest + Predict ===
selected_features = selector.transform(combined_features)
predicted_salary = best_model.predict(selected_features)[0]

print(f"\nüí∞ Predicted Average Salary: ${predicted_salary:.3f}k")
# === Step 10: Pretty Salary Range ===
low_end = predicted_salary * 0.8
high_end = predicted_salary * 1.2

low_formatted = f"${int(round(low_end))}K"
high_formatted = f"${int(round(high_end))}K"

print(f"\nüí∞ Estimated Salary Range: {low_formatted}‚Äì{high_formatted} (Glassdoor est.)")

````
